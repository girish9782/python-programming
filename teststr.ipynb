```python
n=input("enter the string")
i=0
while i<len(n):
    if n[i]=="h":
        print("H")
    elif n[i]=="l":
        print("L")
    elif n[i]=="o":
        print("O")
    elif n[i]=="E":
        print("e")
    elif n[i]=="j":
        print("J")
    elif n[i]=="a":
        print("A")
    elif n[i]=="v":
        print("V")
    else:
        print(n[i])
    i+=1
```
